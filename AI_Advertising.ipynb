```python
import pandas as pd
url = '/content/Advertising.csv'
data = pd.read_csv(url)
data.head()
```

```python
import matplotlib.pyplot as plt
data.hist()
plt.show()
```

```python
correlation = data.corr()
print(correlation)
```

```python
import seaborn as sns
sns.heatmap(correlation,annot=True)
```

```python
sns.pairplot(data)
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['TV']), y= data['sales'])   #fit day du lieu
print('he so intercept la:',model.intercept_)
print('he so coeffient la:',model.coef_)
```

```python
model.score(X=pd.DataFrame(data['TV']), y= data['sales'])
```

```python
y_pred = model.predict(X = pd.DataFrame(data['TV']))
data.plot(kind = 'scatter',x = 'TV',y = 'sales', figsize = (9,9), color = 'black')
plt.plot(data['TV'],y_pred, color = 'blue')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['TV'],data['TV']**2]).T
model.fit(X,y=data['sales'])
print('Intercept cua mo hinh:',model.intercept_)
print('coeffient cua mo hinh:',model.coef_)
print('Do chinh xac cua mo hinh:',model.score(X,y=data['sales']))
```

```python
import numpy as np
range = np.arange(1,300,1)
X_range= pd.DataFrame([range,range**2]).T
y_pred = model.predict(X=X_range)
data.plot(kind = 'scatter',x ='TV',y = 'sales',figsize=(9,9),color = 'black')
plt.plot(range, y_pred,color='blue')
```

```python
model = linear_model.LinearRegression()
X=pd.DataFrame([data['TV'],data['TV']**2,data['TV']**3]).T
model.fit(X,y=data['sales'])
print('Intercept của mô hình:',model.intercept_)
print('Coeffient của mô hình:',model.coef_)
print('Độ chính xác của mô hình:',model.score(X=pd.DataFrame([data['TV'],data['TV']**2,data['TV']**3]).T,y=data['sales']))

import numpy as np
range = np.arange(1,300,1)
X_range=pd.DataFrame([range,range**2,range**3]).T
y_pred = model.predict(X=X_range)
data.plot(kind='scatter',x='TV',y='sales',figsize=(9,9),color='black')
plt.plot(range,y_pred,color='blue')
```
